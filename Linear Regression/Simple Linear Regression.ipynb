```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**Data-Preprocessing**

```python
df = pd.read_csv('/content/salary_data.csv')
df.head()
```

```python
df.shape
```

```python
# Check if there exits any missing value in df
df.isnull().sum()
```

```python
# Split the dataframe into feature and target set
X = df['YearsExperience'].values
Y = df['Salary'].values
```

```python
plt.scatter(X, Y, c='r')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=2)
```

```python
X_train = X_train.reshape(-1, 1)
Y_train = Y_train.reshape(-1, 1)
X_test  = X_test.reshape(-1, 1)
Y_test = Y_test.reshape(-1, 1)
```

```python
regr = LinearRegression()
regr.fit(X_train, Y_train)
```

```python
print(regr.coef_)
print(regr.intercept_)
```

Linear Model = m * x + c

```python
plt.scatter(X_train, Y_train, c='r')
plt.plot(X_train, regr.coef_[0][0] * X_train + regr.intercept_[0], c='b')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Predictions
Y_hat = regr.predict(X_test)
```

```python
# Model Evaluation
from sklearn.metrics import mean_squared_error

print("Mean sqaued error: ", mean_squared_error(Y_test, Y_hat))
```
